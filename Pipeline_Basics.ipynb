```python


# Initialize the NER pipeline
ner_pipeline = pipeline("ner", model="dslim/bert-base-NER", tokenizer="dslim/bert-base-NER")

# Input text
text = "Apple Inc. is an American multinational technology company headquartered in Cupertino, California."

# Perform NER
entities = ner_pipeline(text)

# Print the identified entities and their labels
for entity in entities:
    print(f"Entity: {entity['word']}, Label: {entity['entity']}")
```

```python
from transformers import pipeline

#example: Text Classification
classifier = pipeline("text-classification", model="distilbert-base-uncased-finetuned-sst-2-english")

#input text
text = "I feel extremely happy after receiving positive feedback on my project!"
result = classifier(text)

#print the output
print("Classification Result:", result)
```

### Analysis / Reflection: Text Classification  
**Context of My Prompt:**  
I chose this prompt to test the model's ability to detect positive emotions expressed after a successful project. In real-world applications, sentiment analysis is useful for assessing customer feedback, social media posts, or personal reflections.  

**Result Quality:**  
The output correctly labeled the sentiment as positive, which aligns with the optimistic nature of the prompt. The model's response was accurate and useful for sentiment analysis tasks.  


```python
 from transformers import pipeline

# Custom Example: Text Generation
generator = pipeline("text-generation", model="gpt2")

# Custom prompt
prompt = "The future of technology is driven by innovation and creativity."
output = generator(prompt, max_length=50, num_return_sequences=1)

# Print the generated text
print("Generated Text:", output)
```

### Analysis / Reflection: Text Generation  
**Context of My Prompt:**  
I chose this prompt to explore the model's ability to generate imaginative text about future technology. The idea was to create a short, futuristic narrative that demonstrates the model's creativity and coherence.  

**Result Quality:**  
The generated text was coherent, relevant, and creatively expanded on the given prompt. The model demonstrated a good ability to maintain thematic consistency while introducing new ideas.  


```python
from transformers import pipeline

# Custom Example: Question Answering
qa = pipeline("question-answering", model="distilbert-base-uncased")

# Custom context and question
context = "Albert Einstein developed the theory of relativity, which fundamentally changed the understanding of physics."
question = "Who developed the theory of relativity?"
result = qa(question=question, context=context)

# Print the answer
print("Answer:", result)
```

### Analysis / Reflection: Question Answering  
**Context of My Prompt:**  
I chose this factual prompt to evaluate the model's capacity to extract accurate information from a historical context. Question answering tasks are crucial in applications such as chatbots and automated information retrieval.  

**Result Quality:**  
The output was accurate, as the model correctly identified "Albert Einstein" as the developer of the theory of relativity. This demonstrates that the model can efficiently process and extract factual information from the given context.
